```python
##Dividingn Featuresinto Independent and Dependent
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
##Train test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
## Standardizing the dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
import pickle
pickle.dump(scaler,open("scalling.pkl", "wb"))
```

```python
X_train
```

# Model Training

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
## print Coefficients and intercepts
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
##Parameters model has been trained
regression.get_params()
```

```python
##Predict with test data
reg_pred=regression.predict(X_test)
```


# Assumptions

```python
#plot Scatterplot for prediction
plt.scatter(y_test,reg_pred)
```

```python
##Residuals
residuals=y_test-reg_pred
residuals
```

```python
##plot Residuals
sns.displot(residuals,kind='kde')
```

```python
## Scatter plot with respect to residuals and predictions

## uniform distribution
plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

# R Square and adjusted R Square

Formula
R^2=1-SSR/SST
R^2= Coeffient of determination
SSR=Sum of squares of residuals
SST= Total sum of squares

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```


Adjusted R2=1-[(1-R^2)*(n-1)/(n-k-1)]
Where
R2= R2 of the model
n=Number of Observations
k=Number of predictor variable


```python
#Display adjusted R-Squared
R2=1-(1-score)*(len(y_test)-1)/(len(y_test) - X_test.shape[1]-1)
R2
```
